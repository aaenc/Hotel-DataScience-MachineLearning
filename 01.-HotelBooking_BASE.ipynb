### 7.4.2.- Explicacion

- Método `drop_duplicates()` en Pandas: El método **`drop_duplicates()`** se utiliza para eliminar filas duplicadas de un DataFrame. Es una función muy útil para limpiar datos y asegurarse de que no haya registros repetidos.

- Sintaxis básica
```python
DataFrame.drop_duplicates(
    subset=None,
    keep='first',
    inplace=False,
    ignore_index=False
)

- Parámetros:

    - 1. **`subset`** (opcional)
        - **Tipo**: Lista de columnas o `None`.
        - **Descripción**: Especifica las columnas que se usarán para identificar duplicados.
            - Si no se proporciona (`subset=None`), se consideran todas las columnas del DataFrame.
            - Si se proporciona una lista de columnas, solo se compararán esas columnas para detectar duplicados.
        - **Ejemplo**:
          ```python
          df.drop_duplicates(subset=["columna1", "columna2"])

    - 2. **`keep`** (opcional)
        - **Tipo**: `'first'`, `'last'` o `False`.
        - **Descripción**: Determina cuál de las filas duplicadas se conserva:
            - `'first'` (valor por defecto): Mantiene la primera ocurrencia de cada duplicado y elimina las siguientes.
            - `'last'`: Mantiene la última ocurrencia de cada duplicado y elimina las anteriores.
            - `False`: Elimina todas las filas duplicadas, sin mantener ninguna.
        - **Ejemplo**:
            ```python
              df.drop_duplicates(keep="last")

    - 3. **`inplace`** (opcional)
        - **Tipo**: Booleano (`True` o `False`).
        - **Descripción**: Indica si la operación debe realizarse directamente sobre el DataFrame original.
            - `False` (valor por defecto): Devuelve un nuevo DataFrame sin duplicados.
            - `True`: Modifica el DataFrame original y no devuelve nada.
        - **Ejemplo**:
            ```python
            df.drop_duplicates(inplace=True)

    - 4. `ignore_index` (opcional)
        - **Tipo**: Booleano (`True` o `False`).
        - **Descripción**: Si es `True`, el índice del DataFrame resultante se reinicia (0, 1, 2, ...).
            - `False` (valor por defecto): Mantiene el índice original.
            - `True`: Reinicia el índice después de eliminar duplicados.

### 9.3.3.- Metodo IQR (Rango Intercuartílico):

#### 9.3.3.1.- Explicacion IQR (Rango Intercuartílico):

- **IQR (Rango Intercuartílico)**:  
  Es la diferencia entre el percentil 75 (Q3) y el percentil 25 (Q1). Representa la dispersión del 50% central de los datos.  

  **Fórmula**:  
  `IQR = Q3 - Q1`

- **Límite Inferior**:  
  Es el valor por debajo del cual se considera que un dato es un outlier.  

  **Fórmula**:  
  `Límite Inferior = Q1 - 1.5 * IQR`

- **Límite Superior**:  
  Es el valor por encima del cual se considera que un dato es un outlier.  

  **Fórmula**:  
  `Límite Superior = Q3 + 1.5 * IQR`

- **Interpretación**:
Si un valor en tus datos es **menor que el Límite Inferior** o **mayor que el Límite Superior**, se considera un **outlier**.

#### 9.3.3.2.- Cálculo IQR (Rango Intercuartílico), Limite Inferior y Limite Superior:

````python
# Selecciona solo las columnas numéricas (excluyendo fechas)
#df_numeric = df_original.select_dtypes(include=['float64', 'int64'])
# Seleccionar todas las columnas numéricas
df_numeric = df_original.select_dtypes(include=np.number)

# Calcula las estadísticas descriptivas
desc_stats = df_numeric.describe().T[['mean', 'std', 'min', 'max', '25%', '50%', '75%']]

# Calcula el rango intercuartílico (IQR) y los límites para detectar outliers
desc_stats['IQR'] = desc_stats['75%'] - desc_stats['25%']
desc_stats['Límite Inferior'] = desc_stats['25%'] - 1.5 * desc_stats['IQR']
desc_stats['Límite Superior'] = desc_stats['75%'] + 1.5 * desc_stats['IQR']
````

````python
# Muestra el resultado
display(desc_stats)
````

#### 9.3.3.3.- Cálculamos Outlayers a partir de los Limites Inferior y Superior:

````python
# Calcular la longitud máxima de los nombres de columna
longitud_columna_maxima = max(len(col) for col in df_numeric.columns)
# Vamos a calcula el numero de OUTLAYERS por columna
num_outliers = []
for i in df_numeric.columns:
    outliers = df_numeric[
        (df_numeric[i] < desc_stats.loc[i, 'Límite Inferior']) |
        (df_numeric[i] > desc_stats.loc[i, 'Límite Superior'])
    ]
    num_outliers.append(len(outliers))  # Almacena el número de outliers    
    if len(outliers)>0:
        print(f"   - {i:<{longitud_columna_maxima}}: Tiene {len(outliers)} outlayers.")
    else:
        print(f"   - {i:<{longitud_columna_maxima}}: No Tiene outlayers.")
# Agrega la columna "Número Outliers" al DataFrame desc_stats
desc_stats['Número Outliers'] = num_outliers
````

### 9.3.9.- Calculamos el VIF (Variance Inflation Factor)

#### 9.3.9.1.- Explicación:

- 🔹 ¿Qué es el VIF (Variance Inflation Factor)?

El **VIF (Factor de Inflación de Varianza)** mide la **multicolinealidad** entre las variables de un modelo.  
Sirve para identificar si algunas variables están altamente correlacionadas entre sí, lo que puede afectar la estabilidad del modelo.


#### 9.3.9.2.- ¿Cómo interpretar el VIF?

- 🔹 ¿Cómo interpretar el VIF?

| **VIF**   | **Interpretación**                                      |
|-----------|---------------------------------------------------------|
| **≈ 1**   | No hay multicolinealidad. ✅                            |
| **1 - 5** | Multicolinealidad moderada, puede ser aceptable. ⚠️    |
| **> 5**   | Alta multicolinealidad, puede afectar el modelo. ❌     |
| **> 10**  | Multicolinealidad muy grave, la variable debe eliminarse. 🚨 |
