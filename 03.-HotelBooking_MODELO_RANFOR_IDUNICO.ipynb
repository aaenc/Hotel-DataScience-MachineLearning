# 11.- Normalizar valores
- ¿Por qué es importante este paso?
    - **Consistencia de escala**: Los algoritmos de aprendizaje automático pueden ser sensibles a la escala de los datos. Si las características tienen diferentes escalas (por ejemplo, edad en años y salario en miles de dólares), el modelo podría sesgarse hacia las características con mayor magnitud. El escalado asegura que todas las características tengan la misma importancia.
    - **Mejora el rendimiento**: El escalado puede mejorar la precisión de modelos como K-Nearest Neighbors (KNN), Support Vector Machines (SVM) y redes neuronales, que dependen de distancias entre los puntos de datos.
    - **Evita problemas numéricos**: Las diferencias de escala pueden generar problemas numéricos en el cálculo de ciertos algoritmos. Al estandarizar los datos, se reduce el riesgo de inestabilidad en los cálculos.

    - Esta variable se utiliza para fijar la semilla del generador de números aleatorios (`random_state`), lo que garantiza que la división de los datos sea reproducible. Es decir, cada vez que ejecutes el código, obtendrás la misma división en conjuntos de entrenamiento y prueba.

---

- `train_test_split(X, y, test_size=0.25, random_state=semilla)`:
    - Divide los datos en conjuntos de entrenamiento (`X_train`, `y_train`) y prueba (`X_test`, `y_test`).

- Parámetros:
    - **`X`**: Conjunto de características (variables independientes).
    - **`y`**: Conjunto de etiquetas o valores objetivo (variable dependiente).
    - **`test_size=0.25`**: Especifica que el 25% de los datos se utilizarán para pruebas, y el 75% restante para entrenamiento.
    - **`random_state=semilla`**: Fija la semilla del generador de números aleatorios para garantizar reproducibilidad.

- Resultados:
    - **`X_train`**: Conjunto de características para entrenamiento (75% de los datos).
    - **`X_test`**: Conjunto de características para pruebas (25% de los datos).
    - **`y_train`**: Etiquetas correspondientes a `X_train`.
    - **`y_test`**: Etiquetas correspondientes a `X_test`.

- **¿Por qué es importante `random_state`?**: Al fijar la semilla (`random_state`), se asegura que la división de los datos sea la misma en cada ejecución del código. Esto es crucial para:
    - Reproducir resultados en experimentos.
    - Comparar modelos de manera consistente.
    - Evitar variaciones aleatorias en la evaluación del rendimiento del modelo.

El valor que indica **mucha importancia** en una variable depende del método de medición y del contexto del problema.  

---

- 🔹 1️⃣ Modelos Lineales (Coeficientes Absolutos)
En modelos como **Regresión Logística** o **SVM**, la importancia de una característica se mide con el valor absoluto de su coeficiente (`coef_`).  

- 📌 Regla general:
✔ **Coeficiente grande** → Alta importancia.  
✔ **Coeficiente cercano a 0** → Baja importancia.  

**Ejemplo:**  
| Feature   | Coefficient | Importancia relativa |
|-----------|------------|----------------------|
| Edad      | 2.5        | Alta                |
| Ingresos  | 1.2        | Media               |
| Género    | 0.03       | Baja                |

- 📌 Interpretación:
    - Si un coeficiente absoluto es **mayor que 1**, suele indicar que la variable tiene un **impacto significativo** en la predicción.  
    - Un coeficiente entre **0.1 y 0.5** podría tener una importancia **moderada**.  
    - Valores cercanos a **0** indican que la variable apenas influye.  

⚠ **Nota:** Si las variables no están escaladas (`StandardScaler`), la magnitud del coeficiente dependerá de las unidades de medida.  

---

- 🔹 2️⃣ Modelos Basados en Árboles (`feature_importances_`)
En modelos como **Random Forest**, **XGBoost** o **LightGBM**, la importancia de una variable es un valor entre `0` y `1`, donde `1` significa que la variable es **totalmente determinante**.  

- 📌 Regla general:
✔ **Feature Importance > 0.2** → Variable clave en la predicción.  
✔ **Entre 0.05 y 0.2** → Moderada importancia.  
✔ **Menos de 0.05** → Poca influencia.  

**Ejemplo:**  
| Feature   | Feature Importance |
|-----------|--------------------|
| Edad      | 0.35               |
| Ingresos  | 0.25               |
| Género    | 0.02               |

- 📌 Interpretación:
    - Si una variable tiene **mucha más importancia** que las demás (`> 0.3`), puede ser que el modelo dependa demasiado de ella (**riesgo de sobreajuste**).  
    - Si todas las variables tienen importancias similares, el modelo está distribuyendo el peso entre múltiples factores.  

---

- 🔹 3️⃣ Interpretación con SHAP (Shapley Values)
SHAP mide el impacto de cada variable en cada predicción. No hay un umbral fijo, pero:  

✔ **SHAP alto (positivo o negativo)** → Variable influye mucho en la predicción.  
✔ **SHAP cercano a 0** → La variable apenas contribuye.

### 50.2.1.- Explicacion

- 📊 Matriz de Confusión

Una **matriz de confusión** es una tabla que evalúa el rendimiento de un modelo de clasificación, comparando las predicciones con los valores reales.

- 🏗 **Estructura de la Matriz de Confusión**

| Actual \ Predicho | Predicho **Positivo** (1) | Predicho **Negativo** (0) |
|-------------------|--------------------------|--------------------------|
| **Real Positivo (1)**  | **TP** (Verdadero Positivo)  | **FN** (Falso Negativo)  |
| **Real Negativo (0)**  | **FP** (Falso Positivo)  | **TN** (Verdadero Negativo)  |

---

- 📌 **Definiciones de las Variables**
    - **TP (True Positive - Verdadero Positivo)** 🟩  
      → Casos donde el modelo **predijo "1" y era realmente "1"**  
    - **FP (False Positive - Falso Positivo)** 🔴  
      → Casos donde el modelo **predijo "1" pero era realmente "0"**  
      → También llamado **Error Tipo I**  
    - **FN (False Negative - Falso Negativo)** 🔴  
      → Casos donde el modelo **predijo "0" pero era realmente "1"**  
      → También llamado **Error Tipo II**  
    - **TN (True Negative - Verdadero Negativo)** 🟩  
      → Casos donde el modelo **predijo "0" y era realmente "0"**  

---

- 📈 **Métricas Derivadas**
A partir de la matriz de confusión, se pueden calcular varias métricas importantes:

    - **1️⃣ Precisión (Accuracy)**
        > **¿Qué porcentaje de las predicciones totales son correctas?**  
        \[
        \text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
        \]

    - **2️⃣ Precisión (Precision)**
        > **Cuando el modelo predice "positivo", ¿cuántas veces acierta?**  
        \[
        \text{Precision} = \frac{TP}{TP + FP}
        \]

    - **3️⃣ Sensibilidad (Recall) o Tasa de Verdaderos Positivos**
        > **De todos los casos realmente positivos, ¿cuántos detecta el modelo?**  
        \[
        \text{Recall} = \frac{TP}{TP + FN}
        \]

    - **4️⃣ Especificidad**
        > **De todos los casos realmente negativos, ¿cuántos detecta el modelo?**  
        \[
        \text{Specificity} = \frac{TN}{TN + FP}
        \]

    - **5️⃣ F1-Score**
        > **Media armónica entre precisión y recall (útil en datos desbalanceados)**  
        \[
        F1 = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}
        \]

---

- 🎯 **Ejemplo Numérico**
Supongamos que tenemos la siguiente matriz de confusión:

| Actual \ Predicho | **1 (Positivo)** | **0 (Negativo)** |
|-------------------|-----------------|-----------------|
| **1 (Positivo)**  | 50 (TP)         | 10 (FN)        |
| **0 (Negativo)**  | 5 (FP)          | 100 (TN)       |

🔹 **Accuracy**:  
\[
\frac{50 + 100}{50 + 100 + 5 + 10} = 0.91 \quad (91\%)
\]

🔹 **Precision**:  
\[
\frac{50}{50 + 5} = 0.91 \quad (91\%)
\]

🔹 **Recall**:  
\[
\frac{50}{50 + 10} = 0.83 \quad (83\%)
\]

🔹 **F1-Score**:  
\[
2 \times \frac{0.91 \times 0.83}{0.91 + 0.83} = 0.87 \quad (87\%)
\]

---

- 🏁 **Conclusión**
La **matriz de confusión** es clave para evaluar modelos de clasificación y entender sus errores. Dependiendo del problema, algunas métricas pueden ser más importantes que otras (ej. en diagnóstico médico, el `Recall` es más relevante que la `Precision`).

📌 **Si tienes datos desbalanceados**, la **Precisión (Accuracy) puede ser engañosa**. En su lugar, usa **F1-Score, Precision y Recall** para un análisis más preciso. 🚀

### 50.2.7.- Valor Predictivo Negativo (Negative Predictive Value - NPV)

El **Valor Predictivo Negativo (NPV)** mide la proporción de verdaderos negativos sobre el total de predicciones negativas realizadas por el modelo.  

- 📌 **Fórmula**  
\[
NPV = \frac{TN}{TN + FN}
\]
Donde:  
    - **TN (True Negatives)** → Casos correctamente predichos como negativos.  
    - **FN (False Negatives)** → Casos que eran positivos, pero fueron predichos como negativos.  

- 🔹 **Interpretación**  
    - Un **NPV alto** indica que cuando el modelo predice un resultado negativo, tiene una alta probabilidad de ser correcto.  
    - Un **NPV bajo** sugiere que el modelo está fallando al predecir negativos, clasificando erróneamente muchos casos positivos como negativos (falsos negativos).  

- 📊 **Ejemplo**  
Si un modelo de detección de fraude predice que una transacción **no es fraudulenta**, el **NPV** nos dice con qué confianza podemos creer esa predicción.  

    - Si **NPV = 95%**, significa que **el 95% de las transacciones predichas como "no fraudulentas" realmente no lo son**.  
    - Si **NPV = 60%**, hay una alta probabilidad de que el modelo esté dejando pasar fraudes como si fueran legítimos.  

- ✅ **¿Cuándo es importante el NPV?**  
    - En **medicina** (detección de enfermedades): Si un test dice que un paciente no tiene una enfermedad, el NPV nos indica qué tan confiable es esa afirmación.  
    - En **seguridad** (detección de fraudes o ataques): Un bajo NPV significa que podríamos estar dejando pasar amenazas reales.  

👉 **NPV es clave cuando el costo de un falso negativo es alto.**


````python
vpn=TN / (TN + FN)
print(f"El vpn es: {vpn:.10f}")
````

### 50.2.8.- F1-Score (Balance entre Precisión y Recall)

El F1-Score es la media armónica entre Precisión y Recall. Se usa cuando hay desbalance de clases.

    -📌 Útil cuando hay que equilibrar precisión y recall.

- 📌 Interpretación del F1-Score

El **F1-Score** varía entre **0 y 1**, donde:

| **F1-Score** | **Interpretación** |
|-------------|-----------------|
| **≈ 1** | Excelente equilibrio entre precisión y recall. |
| **0.8 - 0.9** | Buen modelo con buen balance. |
| **0.6 - 0.7** | Modelo aceptable, pero podría mejorar. |

### 50.2.11.- Metricas

````python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score,mean_absolute_percentage_error
#import numpy as np
````

#### Explicacion

- 📌 Métricas clave en modelos de regresión

Las métricas clave son:

✔ **R² (Coeficiente de Determinación)**: Cuánto explica el modelo la varianza en los datos.  
✔ **MAE (Mean Absolute Error)**: Promedio de los errores absolutos.  
✔ **MSE (Mean Squared Error)**: Error cuadrático medio.  
✔ **RMSE (Root Mean Squared Error)**: Raíz cuadrada del MSE (más interpretable).  
✔ **MAPE (Mean Absolute Percentage Error)**: Error porcentual promedio entre las predicciones del modelo y los valores reales.  

#### 50.2.11.1.- **R² (Coeficiente de Determinación)**

#### Explicacion

- El **R² (R cuadrado)** mide qué proporción de la varianza en la variable dependiente es explicada por el modelo. Su valor oscila entre 0 y 1 (o puede ser negativo si el modelo es muy malo).

- 📌 Interpretación de R²  

| **Valor de R²**  | **Interpretación** |
|----------------|------------------------------------------------|
| **≈ 1**  | El modelo explica casi toda la variabilidad de los datos (modelo ideal). |
| **0.8 - 0.9** | Muy buen ajuste del modelo. |
| **0.5 - 0.7** | Modelo aceptable, pero con margen de mejora. |
| **< 0.5** | Modelo con bajo poder explicativo. |
| **< 0 (negativo)** | El modelo es peor que una simple media (modelo inadecuado). |


#### Medida

````python
print("R2 Score:", r2_score(y_test, y_pred))
````

#### 50.2.11.2.- **MAE (Mean Absolute Error)**

#### Explicacion

- 📌 ¿Qué es el MAE (Mean Absolute Error) y cómo se interpreta?  

El **MAE (Mean Absolute Error)** mide el **error promedio absoluto** entre las predicciones del modelo y los valores reales. Es decir, representa la diferencia media en **valor absoluto** entre lo que el modelo predice y la realidad.  

---

- 🔹 **Fórmula del MAE**  
\[
MAE = \frac{1}{n} \sum_{i=1}^{n} | y_i - \hat{y}_i |
\]
Donde:  
- \( y_i \) son los valores reales.  
- \( \hat{y}_i \) son los valores predichos por el modelo.  
- \( n \) es el número total de observaciones.  

---

- 🔹 **¿Cómo interpretar el MAE?**  
- **MAE bajo** → El modelo tiene un error promedio pequeño (mejor rendimiento).  
- **MAE alto** → El modelo comete errores grandes en promedio (peor rendimiento).  
- Se mide en las mismas unidades que la variable objetivo (**y**), lo que facilita su interpretación.  

💡 **Ejemplo:**  
Si estamos prediciendo el precio de casas en **euros (€)** y obtenemos un **MAE = 5,000**, significa que, en promedio, el modelo se equivoca por **5,000€** en cada predicción.  

---

- 📌 **Ventajas y Desventajas del MAE**  
✔ **Ventajas:**  
- Fácil de entender e interpretar.  
- No amplifica los errores grandes (a diferencia del MSE).  

❌ **Desventajas:**  
- No penaliza los errores grandes de manera significativa.  
- No indica si el modelo **subestima** o **sobrestima** (no tiene en cuenta el signo del error).  



#### Medida

````python
print("MAE:", mean_absolute_error(y_test, y_pred))
````

#### 50.2.11.3.- **MSE (Mean Squared Error)**

#### Explicacion

- Error Cuadrático Medio (MSE)

El **Error Cuadrático Medio (MSE)** mide el promedio de los errores al realizar predicciones en un modelo. Este se calcula tomando la media de los cuadrados de las diferencias entre los valores reales y los valores predichos.

- Fórmula del MSE:

$$
MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

Donde:
- \(y_i\): Valor real del dato \(i\).
- \(\hat{y}_i\): Valor predicho por el modelo para el dato \(i\), es decir, el valor estimado del dato \(i\) con el modelo (con el "sombrero" en \(y\)).
- \(n\): Número total de datos.

Un **MSE más bajo** indica un mejor ajuste del modelo.


- Interpretación del MSE

- Un **MSE bajo** indica que el modelo tiene un buen ajuste, ya que los errores promedio son pequeños.
- Un **MSE alto** indica que el modelo está cometiendo errores significativos en las predicciones.
- El **MSE depende de las unidades de los datos**: si las unidades de las variables de salida son grandes, el MSE será grande. Por eso, a veces es más intuitivo usar el **RMSE**, que está en las mismas unidades que los valores reales. 😊


#### Medida

````python
print("MSE:", mean_squared_error(y_test, y_pred))
````

#### 50.2.11.4.- **RMSE (Root Mean Squared Error)**

#### Explicacion

- Fórmula del RMSE

El RMSE (Root Mean Squared Error) se calcula con la siguiente fórmula:

$$
RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}
$$

donde:
- \(n\) es el número total de observaciones.
- \(y_i\) son los valores reales.
- \(\hat{y}_i\) son los valores predichos.

El RMSE mide la diferencia promedio entre los valores reales y los valores predichos, penalizando los errores más grandes debido al cuadrado de las diferencias.


- ¿Qué valor es bueno para el RMSE?

El valor del RMSE (Root Mean Squared Error) por sí solo no tiene una interpretación absoluta, ya que depende del contexto y de la escala de los datos. Sin embargo, a continuación se presentan algunas pautas generales para evaluar si el valor de RMSE es adecuado:

1. **Comparación con la escala de los datos**:
   Un RMSE bajo es generalmente deseable, pero debe ser comparado con el rango de los valores reales (\(y_i\)) o la magnitud de los datos. Por ejemplo:
   - Si los valores reales están en el rango de 0 a 100, un RMSE de 10 puede ser aceptable, pero si los valores van de 0 a 1, un RMSE de 10 sería inaceptable.

2. **Modelos más complejos y datos más ruidosos**:
   En problemas más complejos o cuando los datos son ruidosos, un RMSE más alto puede ser esperado. Sin embargo, si el RMSE es muy alto, podría indicar que el modelo no está capturando bien la relación entre las variables o que hay un problema con los datos.

3. **Mejora continua**:
   El objetivo en muchos casos es minimizar el RMSE lo más posible. A medida que mejoras tu modelo, deberías observar cómo disminuye el RMSE.

4. **Contexto del dominio**:
   Dependiendo del dominio del problema, lo que se considera un RMSE "bueno" puede variar:
   - En aplicaciones de ingeniería de precisión, como la predicción de parámetros físicos o financieros, se podría esperar un RMSE más bajo.
   - En áreas como el análisis de sentimientos o recomendaciones, donde los datos son inherentemente ruidosos, un RMSE relativamente más alto puede ser aceptable.

- Resumen:

- **Un RMSE bajo** es generalmente mejor, pero depende de la escala de los datos.
- **Un valor bueno** depende del contexto, pero en general, un RMSE cercano a 0 indica un buen ajuste del modelo.
  
Una buena referencia sería comparar el RMSE de tu modelo con el de otros modelos o con un modelo base (por ejemplo, uno que predice siempre la media o la mediana de los datos).


#### Medida

````python
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
````

#### 50.2.11.5.- MAPE (Mean Absolute Percentage Error): No usar en valores que contengan CEROS!!!!!

#### Explicacion

- 📌 ¿Qué es el MAPE?

El **MAPE (Mean Absolute Percentage Error)** es una métrica que mide el error porcentual promedio entre las predicciones del modelo y los valores reales. Es útil porque expresa el error en términos de porcentaje, lo que facilita su interpretación.


- 🔹 Posibles causas del **MAPE extremo**  

    - 1️⃣ Valores reales muy cercanos a 0  
Si \( y_i \) tiene valores muy pequeños o **cero**, el denominador se vuelve **muy pequeño o infinito**, generando valores desproporcionados.  

    - 2️⃣ Errores de escala  
Si las predicciones están en una **escala incorrecta** (ej. valores en **millones vs. miles**), el error se magnifica.  

    - 3️⃣ Outliers o datos incorrectos  
Datos anómalos pueden **distorsionar el cálculo** y hacer que el MAPE sea irrealmente alto.  

    - 4️⃣ Datos no normalizados  
Si los datos de entrada no están **bien escalados**, el modelo puede dar predicciones muy erráticas.  


#### Medida

````python
mape = mean_absolute_percentage_error(y_test, y_pred) * 100  # Convertir a porcentaje
print(f"MAPE: {mape:.2f}%")
````

- Filtrar valores reales muy cercanos a 0 antes de calcular el MAPE

````python
y_test_filtered = np.where(y_test == 0, np.nan, y_test)  # Reemplaza 0s por NaN
mape = np.nanmean(np.abs((y_test_filtered - y_pred) / y_test_filtered)) *100
print(f"MAPE corregido: {mape:.2f}%")
print(f"MAPE INV: {100-mape:.2f}%")
````

- 🔹 ¿Cómo interpretar el MAPE?

El resultado se expresa en **porcentaje (%)**, lo que indica el error relativo del modelo respecto a los valores reales.

| **MAPE (%)**  | **Interpretación** |
|--------------|------------------|
| **0 - 10%**  | Excelente predicción |
| **10 - 20%** | Buena precisión |
| **20 - 50%** | Modelo aceptable, pero mejorable |
| **> 50%**    | Modelo con baja precisión (debe mejorarse) |
